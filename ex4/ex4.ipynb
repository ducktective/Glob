## 2. Добавить новую колонку, в которой дата из Expected Start Date конвертирована в дату в формате

###     2.1. dd.mm.YYYY  (назвать DATE_RU)
###     2.2. mm/dd/YYYY  (назвать DATE_EN)

В файле нет такой колонки как Expected Start Date, но есть Estimated Start Date. С ней я и буду работать.

```python
df['DATE_RU'] = df['Estimated Start Date'].copy()
# и не смотря на то, что я сделал копию колонки python мне выдает warning
# не подскажете, что не так?
```

```python
mon_dict = {'авг. 2020':'Август 2020', 
            'апр. 2020': 'Апрель 2020',
            'дек. 2020': "Декабрь 2020",
           "июля 2020": "Июль 2020",
           "июня 2020": "Июнь 2020",
           "марта 2020": "Март 2020",
           "мая 2020": "Май 2020",
           "нояб. 2020": "Ноябрь 2020",
           "окт. 2020": "Октябрь 2020",
           "сент. 2020": "Сентябрь 2020",
           "февр. 2020": "Февраль 2020",
           "янв. 2021": "Январь 2021"}
```

```python
for i in range(df.shape[0]):
    for key, value in mon_dict.items():
        if df['DATE_RU'][i] == key:
            df.loc[i,'DATE_RU'] = value
    
```

```python
for i in range(df.shape[0]):
    #временная строка, чтобы удобней было работать
    temp_date =datetime.strptime(df['DATE_RU'][i], '%B %Y')
    df.loc[i,'DATE_RU'] = temp_date.strftime("%d.%m.%y") 
df
```

```python
df['DATE_EN'] = df['DATE_RU'].copy()
df
```

```python
for i in range(df.shape[0]):
    #временная строка, чтобы удобней было работать
    temp_date =datetime.strptime(df['DATE_EN'][i], '%d.%m.%y')
    df.loc[i,'DATE_EN'] = temp_date.strftime("%m/%d/%y") 
df
```
