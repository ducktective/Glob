```python
# мне очень стыдно за этот блок кода, но ничего лучше я не придумал.
for i in range(df.shape[0]):
    if df['DATE_RU'][i] == 'авг. 2020':
        df['DATE_RU'][i] = 'Август 2020'
    elif df['DATE_RU'][i] == 'апр. 2020':
        df['DATE_RU'][i] = 'Апрель 2020'
    elif df['DATE_RU'][i] == 'дек. 2020':
        df['DATE_RU'][i] = 'Декабрь 2020'
    elif df['DATE_RU'][i] == 'июля 2020':
        df['DATE_RU'][i] = 'Июль 2020'
    elif df['DATE_RU'][i] == 'июня 2020':
        df['DATE_RU'][i] = 'Июнь 2020'
    elif df['DATE_RU'][i] == 'марта 2020':
        df['DATE_RU'][i] = 'Март 2020'
    elif df['DATE_RU'][i] == 'мая 2020':
        df['DATE_RU'][i] = 'Май 2020'
    elif df['DATE_RU'][i] == 'нояб. 2020':
        df['DATE_RU'][i] = 'Ноябрь 2020'
    elif df['DATE_RU'][i] == 'окт. 2020':
        df['DATE_RU'][i] = 'Октябрь 2020'
    elif df['DATE_RU'][i] == 'сент. 2020':
        df['DATE_RU'][i] = 'Сентябрь 2020'
    elif df['DATE_RU'][i] == 'февр. 2020':
        df['DATE_RU'][i] = 'Февраль 2020'
    elif df['DATE_RU'][i] == 'янв. 2021':
        df['DATE_RU'][i] = 'Январь 2021'
```

```python
for i in range(df.shape[0]):
    #временная строка, чтобы удобней было работать
    temp_str = str(datetime.strptime(df['DATE_RU'][i], '%B %Y'))
    df['DATE_RU'][i] = temp_str[8:10] + '.' + temp_str[5:7] + '.' + temp_str[0:4] 
df
```
